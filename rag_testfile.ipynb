# why chunk overiap needed?

This ensures that the text in each chunk shares a portion with the next, preserving semantic continuity.

Overlap helps maintain context across chunk boundaries, reducing hallucinations.

# What happens if chunk size is too large?



 A large chunk may encompass multiple topics, some of which may be relevant to a user query, while others are not. In such a case, the representation of each topic within a single vector can become diluted, which again will affect the retrieval precision.

On the other hand, smaller chunks that maintain a focused context, allow for more precise matching and retrieval of relevant information. By breaking down documents into meaningful segments, the retriever can more accurately locate specific passages or facts, which ultimately improves RAG performance.

# Phase 4 – Vector Storage (Qdrant)

```python
# create embedding model
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from qdrant_client.models import Distance, VectorParams
from langchain_qdrant import QdrantVectorStore
from qdrant_client import QdrantClient

# Setup Qdrant Locally
client = QdrantClient(":memory:")


# Determine Vector Size
vector_size = len(embeddings.embed_query("sample text"))

# Create collection
if not client.collection_exists("test"):
    client.create_collection(
        collection_name="test",
        vectors_config=VectorParams(
            size=vector_size,
            distance=Distance.COSINE
        )
    )

print("Collection created!")
```

```python
# Prepare documents
documents = []
metadatas = []

for i, chunk in enumerate(texts):

    if not chunk.page_content or chunk.page_content.strip() == "":
        continue

    metadata = chunk.metadata.copy()
    metadata["chunk_index"] = i

    documents.append(chunk.page_content)
    metadatas.append(metadata)

# Create vector store
vector_store = QdrantVectorStore(
    client=client,
    collection_name="test",
    embedding=embeddings
)

# Add texts to Qdrant
vector_store.add_texts(
    texts=documents,
    metadatas=metadatas
)

print("Embeddings stored successfully in Qdrant!")
```

# Phase 5 – Retrieval Using LlamaIndex

```python
from llama_index.vector_stores.qdrant import QdrantVectorStore
from llama_index.core import VectorStoreIndex
from llama_index.core import StorageContext
```

```python
# coonecting to already created db

vector_store = QdrantVectorStore(
    client=client,
    collection_name="test"
)
```

```python
vector_store
```

```python
storage_context = StorageContext.from_defaults(
    vector_store=vector_store
)
```

```python
storage_context
```

```python
from llama_index.core import Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

Settings.embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
# create vector index

index = VectorStoreIndex.from_vector_store(
    vector_store=vector_store,
    storage_context=storage_context
)

print("Vector index created successfully!")
```

```python
index
```

```python

Settings.llm = None
query_engine = index.as_query_engine(similarity_top_k=3)
```

```python
query_engine
```

```python
query = "What is 3D MedDiffusion?"

response = query_engine.query(query)

print("Retrieved Chunks:\n")

for node in response.source_nodes:
    print("Score:", node.score)
    print(node.node.text[:300])
    print("-----")
```
